#  Principal Rotation Parameters from DCM

Given the direction cosine matrix (DCM):

$$
[DCM]_{B/N} = 
\begin{bmatrix}
0.925417 & 0.336824 & 0.175365 \\
0.029696 & -0.521281 & 0.852869 \\
0.377786 & -0.784102 & -0.492404
\end{bmatrix}
$$

Find the equivalent principal rotation parameter set,

$$
\{\hat{\mathbf{e}}, \Phi\}
$$



```python
import numpy as np

BN = np.array([ [0.925417, 0.336824,   0.173648],
                [0.0296956, -0.521281, 0.852869],
                [0.377786, -0.784102, -0.492404] ])

phi = np.arccos(np.clip(0.5 * (np.trace(BN) - 1), -1.0, 1.0))

if np.isclose(phi, np.pi):
    print("Singualarity caused by 2 * sin(pi) in the denominator of e_hat equation.")
    e = np.array([
        np.sqrt((BN[0,0] + 1)/2),
        np.sqrt((BN[1,1] + 1)/2),
        np.sqrt((BN[2,2] + 1)/2)
    ])
else:
    e = 1/(2*np.sin(phi)) * np.array([
        BN[1,2] - BN[2,1],
        BN[2,0] - BN[0,2],
        BN[0,1] - BN[1,0]
    ])

print(f"e={e}, phi={phi}")
```
