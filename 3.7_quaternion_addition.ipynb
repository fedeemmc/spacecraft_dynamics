```python
import numpy as np

beta_d = np.array([0.774597, 0.258199, 0.516398, 0.258199])
beta_dd = np.array([0.359211, 0.898027, 0.179605, 0.179605])

b0, b1, b2, b3 = beta_dd
c0, c1, c2, c3 = beta_d


B_dd = np.array([[b0, -b1, -b2, -b3],
                 [b1, b0, b3, -b2],
                 [b2, -b3, b0, b1],
                 [b3, b2, -b1, b0]])

B_d = np.array([[c0], [c1], [c2], [c3]])

Beta = B_dd @ B_d

if Beta[0,0] < 0:
    Beta = -Beta
print(Beta)

#print(B_dd@B_dd.T)
```
