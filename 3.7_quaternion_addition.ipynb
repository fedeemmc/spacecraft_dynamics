```python
import numpy as np

beta_BN = np.array([0.774597, 0.258199, 0.516398, 0.258199]) # Beta'
beta_FB = np.array([0.359211, 0.898027, 0.179605, 0.179605]) # Beta''


b0, b1, b2, b3 = beta_FB
c0, c1, c2, c3 = beta_BN

beta = np.array([
    b0*c0 - b1*c1 - b2*c2 - b3*c3,
    b0*c1 + b1*c0 + b2*c3 - b3*c2,
    b0*c2 - b1*c3 + b2*c0 + b3*c1,
    b0*c3 + b1*c2 - b2*c1 + b3*c0
])

print(beta)
print(np.round(beta @ beta))
```

```python
import numpy as np

def quat_compose_schaub(beta_dd, beta_d):
    """
    Schaub quaternion composition (scalar-first).
    beta_dd âŠ• beta_d
    """
    b0, b1, b2, b3 = beta_dd
    c0, c1, c2, c3 = beta_d

    return np.array([
        b0*c0 - b1*c1 - b2*c2 - b3*c3,
        b0*c1 + b1*c0 + b2*c3 - b3*c2,
        b0*c2 - b1*c3 + b2*c0 + b3*c1,
        b0*c3 + b1*c2 - b2*c1 + b3*c0
    ])

def quat_conjugate(beta):
    """Inverse of a unit quaternion"""
    return np.array([beta[0], -beta[1], -beta[2], -beta[3]])

def normalize(beta):
    return beta / np.linalg.norm(beta)

def short_rotation(beta):
    """Ensure short rotation (beta0 >= 0)"""
    return beta if beta[0] >= 0 else -beta



beta_FN = np.array([0.359211, 0.898027, 0.179605, 0.179605])
beta_BN = np.array([-0.377964, 0.755929, 0.377964, 0.377964])

beta_NB = quat_conjugate(beta_BN)
beta_FB = quat_compose_schaub(beta_FN, beta_NB)
beta_FB = normalize(beta_FB)
beta_FB = short_rotation(beta_FB)

print("beta_F/B =", np.round(beta_FB, decimals=4))


```
