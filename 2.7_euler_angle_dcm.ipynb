```python
import numpy as np

theta_1 = np.deg2rad(20)
theta_2 = np.deg2rad(10)
theta_3 = np.deg2rad(-10)


# 1st axis rotation (Roll)
C1 = np.array([
    [1, 0,                              0],
    [0, np.cos(theta_3),  np.sin(theta_3)],
    [0, -np.sin(theta_3), np.cos(theta_3)]
])

# 2nd axis rotation (Pitch)
C2 = np.array([
    [np.cos(theta_2), 0, -np.sin(theta_2)],
    [0,               1,                0],
    [np.sin(theta_2), 0,  np.cos(theta_2)]
])

# 3rd axis rotation (Yaw)
C3 = np.array([
    [np.cos(theta_1),  np.sin(theta_1), 0],
    [-np.sin(theta_1), np.cos(theta_1), 0],
    [0,                0,               1]
])


C = C1 @ C2 @ C3

print(np.allclose(C @ C.T, np.eye(3)))

print(np.linalg.det(C))

# 3-1-3 Euler angles extracted FROM THE DCM
Omega = np.arctan2(C[2,0], -C[2,1]) 
i     = np.arccos(C[2,2])
omega = np.arctan2(C[0,2], C[1,2])

print(f"{Omega}, {i}, {omega}")
```

```python
import sympy as sp

# SÃ­mbolos
phi, L, r = sp.symbols('phi L r', real=True)

theta = L*phi/r

# 3rd axis rotation (n3)
R_n3 = sp.Matrix([
    [sp.cos(phi),  sp.sin(phi), 0],
    [-sp.sin(phi), sp.cos(phi), 0],
    [0,             0,          1]
])

# 1st axis rotation (bL)
R_bL = sp.Matrix([[1,            0,           0],
                  [0, sp.cos(-theta),  sp.sin(-theta)],
                  [0, -sp.sin(-theta), sp.cos(-theta)]]   )

# DCM BN
BN = sp.simplify(R_bL * R_n3)

BN
```
